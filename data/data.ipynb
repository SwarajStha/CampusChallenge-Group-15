```python
import pandas as pd

daily_return_data = pd.read_csv('train_return_data_daily.csv')

# Check column names first
print("Columns:", daily_return_data.columns.tolist())

# Convert date column to datetime
daily_return_data['date'] = pd.to_datetime(daily_return_data['date'])

# Filter for 2024: 2024-07-01 to 2024-12-31
daily_return_data_filtered = daily_return_data[(daily_return_data['date'] >= '2024-07-01') & (daily_return_data['date'] <= '2024-12-31')]

daily_return_data_filtered
```
