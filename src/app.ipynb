# CSV previews
Display the first few rows of the workspace CSV files.

```python
import pandas as pd
from pathlib import Path

base_path = Path('.')
train_df_path = base_path / 'train_df.csv'
train_return_data_path = base_path / 'train_return_data.csv'

pd.set_option('display.max_colwidth', 120)

train_df = pd.read_csv(train_df_path)
train_return_data = pd.read_csv(train_return_data_path)

print('train_df head:')
display(train_df.head())

print('\ntrain_return_data head:')
display(train_return_data.head())
```

```python
train_df_tz_naive = train_df.copy()
train_df_tz_naive["date"] = pd.to_datetime(train_df_tz_naive["date"], utc=True, errors="coerce").dt.tz_localize(None)

train_return_dates = pd.to_datetime(train_return_data["date"], errors="coerce")
min_train_df_date = train_df_tz_naive["date"].min()
filtered_return_data = train_return_data.loc[train_return_dates >= min_train_df_date].copy()

print(f"Earliest train_df date (tz-naive): {min_train_df_date}")
print("train_df head (tz-naive dates):")
display(train_df_tz_naive.head())

print("\nFiltered train_return_data head (date >= earliest train_df date):")
display(filtered_return_data.head())
```

```python
from pathlib import Path
import pandas as pd

def aggregate_news_by_return(news_df: pd.DataFrame, returns_df: pd.DataFrame) -> pd.DataFrame:
    news_by_ticker = dict(tuple(news_df.groupby('ticker')))
    agg_rows = []

    for ticker, ret_group in returns_df.groupby('TICKER'):
        if ticker not in news_by_ticker:
            continue
        news_subset = news_by_ticker[ticker].copy()
        ret_sorted = ret_group.sort_values('date').copy()
        ret_sorted['start'] = ret_sorted['date'].shift(1)
        ret_sorted['start'] = ret_sorted['start'].fillna(ret_sorted['date'])

        intervals = pd.IntervalIndex.from_arrays(ret_sorted['start'], ret_sorted['date'], closed='right')
        idx = intervals.get_indexer(news_subset['date'])
        news_subset = news_subset.loc[idx != -1].copy()
        news_subset['return_date'] = ret_sorted.iloc[idx[idx != -1]]['date'].values

        grouped = (
            news_subset.groupby('return_date', group_keys=False)[['date', 'title']]
            .apply(
                lambda g: pd.Series(
                    {
                        'news_count': len(g),
                        'titles_joined': ' | '.join(g.sort_values('date')['title'].tolist()),
                    }
                )
            )
            .reset_index()
        )

        for _, row in grouped.iterrows():
            agg_rows.append(
                {
                    'ticker': ticker,
                    'return_date': row['return_date'],
                    'news_count': int(row['news_count']),
                    'titles_joined': row['titles_joined'],
                }
            )

    return pd.DataFrame(agg_rows)

base_path = Path('.')
news_df = pd.read_csv(base_path / 'train_df.csv')
returns_df = pd.read_csv(base_path / 'train_return_data.csv')

news_df['date'] = pd.to_datetime(news_df['date'], utc=True, errors='coerce').dt.tz_localize(None)
returns_df['date'] = pd.to_datetime(returns_df['date'], errors='coerce')

aggregated_news = aggregate_news_by_return(news_df, returns_df)
display(aggregated_news.head())
```
